**3) Stick Figures**

```python
import matplotlib.pyplot as plt

joints = {
    'head': (0, 3),
    'neck': (0, 2),
    'left_shoulder': (-1, 2),
    'left_elbow': (-2, 1),
    'left_hand': (-3, 1),
    'right_shoulder': (1, 2),
    'right_elbow': (2, 1),
    'right_hand': (3, 1),
    'torso_top': (0, 2),
    'torso_bottom': (0, 0),
}

limbs = [('head', 'neck'), ('neck', 'left_shoulder'), ('neck', 'right_shoulder'),
         ('left_shoulder', 'left_elbow'), ('left_elbow', 'left_hand'),
         ('right_shoulder', 'right_elbow'), ('right_elbow', 'right_hand'),
         ('neck', 'torso_top'), ('torso_top', 'torso_bottom')]

for joint1, joint2 in limbs:
    x = [joints[joint1][0], joints[joint2][0]]
    y = [joints[joint1][1], joints[joint2][1]]
    plt.plot(x, y, marker='o')

plt.xlim(-4, 4)
plt.ylim(-1, 4)
plt.gca().invert_yaxis()
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.show()
```
